# <div align="center">Pie Chart</dev>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Unilities"]
```

```python
# plt.axis = 'equal' ## run if the pie seems elliptical
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
## if we want to increase the size/ radius

plt.pie(exp_vals, labels=exp_labels, radius=2)
plt.show()
```

```python
## if we want to see the percentages

plt.pie(exp_vals, labels=exp_labels, radius=2, autopct="%0.2f%%")
plt.show()
```

```python
## if we want to see the shadow

plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", shadow=True)
plt.show()
```

```python
## if we want to take out the pies

plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", 
        shadow=True, explode=[0,0.2,0,0,0.5]) ## add 0's as many as there are pies, then increase the values
plt.show()
```

```python
## if we want to change the angle of pie

plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", 
        shadow=True, explode=[0,0.2,0,0,0.5], startangle=45) ## add 0's as many as there are pies, then increase the values
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", 
        shadow=True, explode=[0,0.2,0,0,0.5], startangle=180) ## add 0's as many as there are pies, then increase the values
plt.show()
```
