```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y, 'g+--')
```

```python
## Order doesnt matter

plt.plot(x,y, 'r+-.')
```

```python
plt.plot(x,y, 'rD')
```

```python
plt.plot(x,y, color='red', marker='D', linestyle='')
```

```python
plt.plot(x,y, color='#444455', marker='+', linestyle='', markersize=20)
```

```python
## alpha controls the opacity

plt.plot(x,y, color='green', alpha=0.2)
```
