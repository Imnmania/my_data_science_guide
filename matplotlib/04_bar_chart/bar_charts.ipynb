```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.bar(ypos,revenue)
```

```python
plt.bar(company,revenue)
```

```python
## in the new version we no longer need to import np or xticks
plt.xticks(ypos,company)
plt.bar(ypos,revenue)
```

```python
# plt.xlabel('Company')
plt.title('Us Tech Stocks')
plt.ylabel('Revenue')
plt.bar(company,revenue, label="Revenue")
plt.legend()
```

### Multiple Bar Charts

```python
## For multiple bars or x values
# lets say we want to see the profit and revenue altogether

profit=[40,2,34,12]
```

```python
plt.title('Us Tech Stocks')
plt.ylabel('Revenue')
plt.bar(company,revenue, label="Revenue")
plt.bar(company,profit, label="Profit")
plt.legend()
```

```python
## we want to see them side by side
ypos = np.arange(len(company))
plt.title('Us Tech Stocks')
plt.ylabel('Revenue')

## we make some gap between them by adding and subbing the same number
## and by adding width of same amount
plt.bar(ypos - 0.2,revenue , width=0.4, label="Revenue")
plt.bar(ypos + 0.2,profit , width=0.4, label="Profit")
plt.legend()
```

### Horizontal Bar Charts

```python
plt.yticks(ypos, company)
ypos = np.arange(len(company))
plt.title('Us Tech Stocks')
plt.xlabel('Revenue')

## barh mean horizontal bar chart

plt.barh(ypos - 0.2,revenue, label="Revenue")
plt.barh(ypos + 0.2,profit, label="Profit")
plt.legend()
```
