# Distplots
---
### Distribution Plots
---

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

### Load the Penguins dataset

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

### Drawing a simple DistPlot

```python
sns.distplot(penguins.flipper_length_mm)
```

```python
## we can see Kernel Density (KDE) and Histogram here
## to see rug plots (all parameters) we use "rug" parameter

sns.distplot(penguins.flipper_length_mm, rug=True)
```

### Removing the KDE plot

```python
sns.distplot(penguins.flipper_length_mm, kde = False)
```

### Removing the Histogram

```python
sns.distplot(penguins.flipper_length_mm, hist = False)
```

### Change the number of bins (default 10)

```python
sns.distplot(penguins.flipper_length_mm, bins=20)
```

### Setting bin values as array

```python
sns.distplot(penguins.flipper_length_mm, bins = [160,170,180,190,200,250])
```

### Fitting the data to a particular distribution using norm_hist

```python
## not recommended

sns.distplot(penguins.flipper_length_mm, norm_hist=True, kde=False)
```

### Styling RugPlots

```python
## kws == keywords

sns.distplot(penguins.flipper_length_mm, rug=True, rug_kws = {
    'color': 'red',
    'height': 0.30
})
```

### Styling Histogram

```python
sns.distplot(penguins.flipper_length_mm, hist_kws = {
    'color': 'blue',
    'alpha': 0.3,
    'histtype': 'step'
}, kde = False)
```

```python
sns.distplot(penguins.flipper_length_mm, hist_kws = {
    'color': 'blue',
    'alpha': 0.7,
    'histtype': 'stepfilled'
}, kde = False)
```

### Styling KDE plots

```python
sns.distplot(penguins.flipper_length_mm, kde = True, hist = False, kde_kws = {
    'color': 'green',
    'lw': 4,
})
```
